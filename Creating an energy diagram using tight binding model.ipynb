```python
import matplotlib.pyplot as plt
from energydiagram import ED
import numpy as np
from numpy.linalg import eig
```

# Pure Sb

```python
d = 2.9;
Ess = -14.8;
Epp = -7.24;
Vssσ = -1.4*7.62/(d**2);
Vspσ = 1.84*7.62/(d**2);
Vppσ = 3.1*7.62/(d**2);
Vppπ = -0.63*7.62/(d**2);
```

No s-p interaction

```python
hamiltonian = np.array(
           [[Ess, Vssσ, 0, 0],
            [Vssσ, Ess, 0, 0],
            [0, 0, Epp, Vppσ],
            [0, 0, Vppσ, Epp]])

wn, v = np.linalg.eig(hamiltonian)

print("Eigenvalues:")
display(wn)

#print("\nEigenvectors:")
#display(v)

a = ED()

a.add_level(Ess,'s',top_text='')
a.add_level(Epp,'p','last',top_text='')
a.add_level(wn[0],'$\sigma*$',top_text='')
a.add_level(wn[1],'$\sigma$','last',top_text='')
a.add_level(wn[2],'$\sigma*$','last',top_text='')
a.add_level(wn[3],'$\sigma$','last',top_text='')
a.add_level(Ess,'s',top_text='')
a.add_level(Epp,'p','last',top_text='')

for i in range(2,4):
    a.add_link(0,i,color='g')
    a.add_link(i,6,color='g')
    
for i in range(4,6):
    a.add_link(1,i,color='b')
    a.add_link(i,7,color='b')

a.plot(show_IDs=False)
plt.show()
```

With s-p interaction

```python
hamiltonian = np.array(
           [[Ess, Vssσ, 0, Vspσ],
            [Vssσ, Ess, -Vspσ, 0],
            [0, -Vspσ, Epp, Vppσ],
            [Vspσ, 0, Vppσ, Epp]])

w, v = np.linalg.eig(hamiltonian)

print("Eigenvalues:")
display(w)

#print("\nEigenvectors:")
#display(v)

a = ED()

a.add_level(Ess,'s',top_text='')
a.add_level(Epp,'p','last',top_text='')
a.add_level(w[0],'$\sigma*$',top_text='')
a.add_level(w[1],'$\sigma$','last',top_text='')
a.add_level(w[2],'$\sigma*$','last',top_text='')
a.add_level(w[3],'$\sigma$','last',top_text='')
a.add_level(Ess,'s',top_text='')
a.add_level(Epp,'p','last',top_text='')

for i in range(2,6):
    a.add_link(0,i,color='g')
    a.add_link(i,6,color='g')
    a.add_link(1,i,color='b')
    a.add_link(i,7,color='b')
    
    
a.plot(show_IDs=False)
plt.show()
```

```python
hamiltonian = np.array(
           [[Ess, Vssσ, 0, Vspσ, 0, 0],
            [Vssσ, Ess, -Vspσ, 0, 0, 0],
            [0, -Vspσ, Epp, Vppσ, 0, 0],
            [Vspσ, 0, Vppσ, Epp, 0, 0],
            [0, 0, 0, 0, Epp, Vppπ],
            [0, 0, 0, 0, Vppπ, Epp]])

w, v = np.linalg.eig(hamiltonian)

print("Eigenvalues:")
display(w)

#print("\nEigenvectors:")
#display(v)

a = ED()

a.add_level(Epp,'p',top_text='')
a.add_level(Epp,'p',top_text='')
a.add_level(Ess,'s','last',top_text='')
a.add_level(w[0],'$\sigma*$',top_text='')
a.add_level(w[1],'$\sigma$','last',top_text='')
a.add_level(w[2],'$\sigma*$','last',top_text='')
a.add_level(w[3],'$\sigma$','last',top_text='')
a.add_level(w[4],'$\pi*$','last',top_text='')
a.add_level(w[5],'$\pi$','last',top_text='')
a.add_level(Ess,'s',top_text='')
a.add_level(Epp,'p','last',top_text='')
a.add_level(Epp,'p',top_text='')

for i in range(3,7):
    a.add_link(1,i,color='g')
    a.add_link(i,10,color='g')
    
for i in range(3,7):
    a.add_link(2,i,color='b')
    a.add_link(i,9,color='b')
    
for i in range(7,9):
    a.add_link(1,i,color='r')
    a.add_link(i,10,color='r')

a.plot(show_IDs=False)
plt.show()
```
